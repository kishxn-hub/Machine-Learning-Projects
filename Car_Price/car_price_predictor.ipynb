Model

```python
X= car.drop(columns='Price')       #X has all attributes except price
y= car['Price']
```

```python
#without random state
from sklearn.model_selection import train_test_split           
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder #converts alpha bet num
from sklearn.compose import make_column_transformer  #transform the columns
from sklearn.pipeline import make_pipeline  #Sending rawdata to 1 end & output from another
```

```python
ohe = OneHotEncoder()    #Object of onehotencoder and converts alphbtcl data to numrical
ohe.fit(X[['name','company','fuel_type']])       #converts to num
```

```python
ohe.categories_                 #all values under each column specified
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']), remainder='passthrough')
#making 2 diff operations together to send into pipeline
```

```python
lr = LinearRegression()  #used to predict the value of a variable based on the value of another variable
```

```python
pipe=make_pipeline(column_trans,lr)  #converting values and sending to model
```

```python
pipe.fit(X_train,y_train)  #training the data
```

```python
#training the X_train , y_train data and predicting output from X_test(already trained data) & storing in variable(y_pred) then
#checking for the r2_score using y_test and predicted output (y_pred)
```

```python
y_pred=pipe.predict(X_test)         #prediction value (money)  #X_test is the set of new data used for prediction
```

```python
y_pred                             #predicted money for each data row
```

```python
r2_score(y_test,y_pred)                #r2 score ,first test value passed then predicted
```

```python
scores=[]
for i in range(1000):
    X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.4, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    print(r2_score(y_test,y_pred), i)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)  #returns indices of the max element of the array in a particular axis
```

```python
scores[419]  #to print particular value score[index]
```

```python
#with random state
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.4, random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('LinearRegressionModel.pkl','wb'))
```
